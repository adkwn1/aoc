```python
from math import inf

#filename = './input.txt'
filename = './test.txt'

with open(filename, 'r') as f:
    grid = [line.strip() for line in f.readlines()]

row = 0
for r in grid:
    if 'S' in r:
        start = (row, r.index('S'))
    if 'E' in r:
        end = (row, r.index('E'))
    row += 1

steps = {0: (-1, 0),
         90: (0, 1),
         180: (1, 0),
         270: (0, -1)}

seen = set()
def bfs(start, end):
    queue = []
    queue.append((start, 90, 0))
    min_cost = inf
    while queue:
        (r, c), drx, cost = queue.pop(0)
        if (r, c) == end:
            if cost < min_cost:
                min_cost = cost
            continue
        elif (r, c, drx) in seen:
            print(r, c, queue)
            #if cost >= seen.get((r, c, drx)):
            continue
        else:
            # straight
            nr, nc = steps.get(drx)
            if grid[r + nr][c + nc] != '#':
                queue.append(((r+nr, c+nc), drx, cost + 1 ))
            # turn left
            left = (drx - 90) % 360
            nr, nc = steps.get(left)
            if grid[r + nr][c + nc] != '#':
                queue.append(((r+nr, c+nc), left, cost + 1001 ))
            # turn right
            right = (drx + 90) % 360
            nr, nc = steps.get(right)
            if grid[r + nr][c + nc] != '#':
                queue.append(((r+nr, c+nc), right, cost + 1001 ))
            
            #seen.update({(r, c): cost})
            seen.add((r, c, drx))
            
    return min_cost


print(bfs(start, end))

# 91404 too high
```
