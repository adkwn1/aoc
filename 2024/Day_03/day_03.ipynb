```python
import re

filename = 'input.txt'
pattern = "(mul\(\d{1,3},\d{1,3}\))"

with open(filename, 'r') as f:
    muls = re.findall(pattern, f.read())

total = 0
for m in muls:
    nums = re.findall("\d+", m)
    total += int(nums[0]) * int(nums[1])

print(total)
```

```python
import re

filename = 'input.txt'
pattern = r"(don't\(\))|(do\(\))|(mul\(\d{1,3},\d{1,3}\))"

with open(filename, 'r') as f:
    result = re.findall(pattern, f.read())

do = True
total = 0
for op in result:
    if op[0]:
        do = False
        continue
    elif op[1]:
        do = True
        continue
    else:  
        if do:
            nums = re.findall("\d+", op[2])
            total += int(nums[0]) * int(nums[1])

print(total)
```
