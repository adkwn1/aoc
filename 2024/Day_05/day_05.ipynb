```python
#filename = 'test.txt'
filename = 'input.txt'
  
pg_order = {}
updates = []
switch = False
with open(filename, 'r') as f:
    for line in f:
        if line == '\n':
            switch = True
            continue
        if switch:
            l = line.strip().split(',')
            updates.append(l)
        else:
            l = line.strip().split('|')
            if l[0] in pg_order:
                pg_order[l[0]].add(l[1])
            else:
                pg_order[l[0]] = {l[1]}
                

def checkRules(update:list):
    seen = set()
    mid = len(update)//2
    fixed = []
    changed = False
    for pg in update:
        seen.add(pg)
        rules = pg_order.get(pg, set())
        if (seen & rules):
            changed = True
            i = len(fixed)-1
            for elem in list(seen & rules):
                idx = fixed.index(elem)
                if idx < i:
                    i = idx
            fixed.insert(i, pg)
        else:
            fixed.append(pg)
    
    if changed:
        return int(fixed[mid])
    else:
        return 0

total = 0
for update in updates:
    total += checkRules(update)

print(total)
```
