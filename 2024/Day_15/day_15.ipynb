```python
import re

filename = './input.txt'
#filename = './test.txt'

with open(filename, 'r') as f:
    map, drx = f.read().split('\n\n')

map = map.split('\n')
drx = re.sub(f'\n', '', drx)
num_r, num_c = len(map), len(map[0])
num_d = len(drx)

grid = []
for r in range(num_r):
    row = []
    for c in range(num_c):        
        if map[r][c] == '@':
            row += ['@', '.']
            fish = (r, 2*c)
        elif map[r][c] == 'O':
            row += ['[', ']']
        else:
            row.append(map[r][c])
            row.append(map[r][c])            
    grid.append(row)

moves = {'<': (0, -1),
         '^': (-1, 0),
         '>': (0, 1),
         'v': (1, 0)}

def pushBoxesUD(drx, r, c):
    fr = r
    nr, nc = moves.get(drx)
    boxes = {}
    if grid[r][c] == ']':
        boxes.update({r: {(c-1, c)}})
    else:
        boxes.update({r: {(c, c+1)}})

    shouldMove = False
    while not shouldMove:
        boxes[r + nr] = set()
        shouldMove = True
        for cols in boxes.get(r):
            left, right = cols
            if grid[r + nr][left] == '#' or grid[r + nr][right] == '#':
                return False
            else:
                if grid[r + nr][left] == ']':
                    boxes[r + nr].add((left-1, left))
                    shouldMove = False
                if grid[r + nr][right] == '[':
                    boxes[r + nr].add((right, right+1))
                    shouldMove = False
                if grid[r + nr][left] == '[' or grid[r + nr][right] == ']':
                    boxes[r + nr].add((left, right))
                    shouldMove = False
        r += nr

    while r != fr:
        for col in boxes.get(r - nr):
            left, right = col
            grid[r][left] = '['
            grid[r][right] = ']'
            grid[r - nr][left] = '.'
            grid[r - nr][right] = '.'
        r -= nr
    return True

def pushBoxesLR(drx, r, c):
    fc = c
    nr, nc = moves.get(drx)
    while grid[r][c] in '[]':
        c += 2*nc

    if grid[r][c] == '.':
        while c != fc:
            grid[r][c] = grid[r][c-nc]
            c -= nc
        return True
    else:
        return False

r, c, = fish
for i in range(num_d):
    nr, nc = moves.get(drx[i])
    if grid[r + nr][c + nc] == '.':
        grid[r][c] = '.'
        r += nr
        c += nc
        grid[r][c] = '@'
    elif grid[r + nr][c + nc] in "[]":
        if drx[i] in '^v':
            if pushBoxesUD(drx[i], r + nr, c):
                grid[r][c] = '.'
                r += nr
                grid[r][c] = '@'
        else:
            if pushBoxesLR(drx[i], r, c + nc):
                grid[r][c] = '.'
                c += nc
                grid[r][c] = '@'
    fish = (r, c)
    
total = 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == '[':
            total += (100*i+j)

print(total)
for row in grid:
    print(''.join(row))
```
