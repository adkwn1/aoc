```python
import sys
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

sys.setrecursionlimit(int(10e7))

filename = './input.txt'
#filename = './test.txt'

LAN = nx.Graph()
tnodes = set()
with open(filename, 'r') as f:
    for line in f.readlines():
        u, v = line.strip().split('-')
        LAN.add_edge(u, v)
        if 't' == u[0]:
            tnodes.add(u)
        if 't' == v[0]:
            tnodes.add(v)

count = set()
for c in nx.find_cliques(LAN):
    if tnodes & set(c) and len(c) >= 3:
        for tx in (tnodes & set(c)):
            x = set(c)
            x.remove(tx)
            for pair in combinations(x, 2):
                t = list(pair)
                t.append(tx)
                count.add(', '.join(sorted(t)))                

print(len(count))
```

```python
print(','.join(sorted(max(nx.find_cliques(LAN), key=len))))
```

```python
pos = nx.spring_layout(LAN)
nx.draw(LAN, pos, with_labels=True, node_color='skyblue', edge_color='gray')
plt.show()
```
