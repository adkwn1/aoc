```python
#filename = 'test.txt'
filename = 'input.txt'

#So, a report only counts as safe if both of the following are true:
    #The levels are either all increasing or all decreasing.
    #Any two adjacent levels differ by at least one and at most three.
def checker(report:list[str]) -> int:
    length = len(report)   
    diff = int(report[0]) - int(report[1])
    
    if (diff < 0) and (-3 <= diff <= -1):
        l = -3
        r = -1
    elif (diff > 0) and (1 <= diff <= 3):
        l = 1
        r = 3
    else:
        return 0
    
    idx = 1
    while idx < length-1:
        diff = int(report[idx]) - int(report[idx+1])
        if (l <= diff <= r):
            idx += 1
        else:
            return 0
        
    return 1            
        
        
count = 0
with open(filename, 'r') as f:
    for line in f:
        count += checker(line.split())

print(count)
```

```python
#filename = 'test.txt'
filename = 'input.txt'

#So, a report only counts as safe if both of the following are true:
    #The levels are either all increasing or all decreasing.
    #Any two adjacent levels differ by at least one and at most three.
def checkUnsafe(report:list[str]) -> bool:
    length = len(report)   
    diff = int(report[0]) - int(report[1])
    
    if (diff < 0) and (-3 <= diff <= -1):
        l = -3
        r = -1
    elif (diff > 0) and (1 <= diff <= 3):
        l = 1
        r = 3
    else:
        return True
    
    idx = 1
    while idx < length-1:
        diff = int(report[idx]) - int(report[idx+1])
        if (l <= diff <= r):
            idx += 1
        else:
            return True
        
    return False            
        

def dampner(report:list[str]) -> bool:
    for i in range(len(report)):
        if not (checkUnsafe(report[:i] + report[i+1:])):
            return True

    return False


count = 0
with open(filename, 'r') as f:
    for line in f:
        if checkUnsafe(line.split()):
            if dampner(line.split()):
                count += 1
        else:
            count += 1

print(count)
```
