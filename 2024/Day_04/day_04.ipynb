```python
#filename = 'test.txt'
filename = 'input.txt'

word = {'X': 'M',
        'M': 'A',
        'A': 'S'}

dirMap = {7: (-1, -1),
          8: (-1, 0),
          9: (-1, 1),
          4: (0, -1),
          6: (0, 1),
          1: (1, -1),
          2: (1, 0),
          3: (1, 1)}

grid = []
with open(filename, 'r') as f:
    for line in f:
        grid.append(list(line.strip()))

num_r = len(grid)
num_c = len(grid[0])

###

def inBounds(r, c) -> bool:
    return (0 <= r < num_r) and (0 <= c < num_c)

def traverse(r:int, c:int, curr:str, direction:int):
    if (curr == grid[r][c] == 'S'):
        return 1
    elif grid[r][c] != curr:
        return 0
    else:
        a, b = dirMap.get(direction)
        nxt = word.get(curr)
        r += a
        c += b
        if inBounds(r, c):
            return traverse(r, c, nxt, direction)
        else:
            # OOB
            return 0

def start(r:int, c:int):
    count = 0
    if inBounds(r-1, c-1):
        count += traverse(r-1, c-1, 'M', 7)
    if inBounds(r-1, c):
        count += traverse(r-1, c, 'M', 8)
    if inBounds(r-1, c+1):
        count += traverse(r-1, c+1, 'M', 9)
    if inBounds(r, c-1):
        count += traverse(r, c-1, 'M', 4)
    if inBounds(r, c+1):
        count += traverse(r, c+1, 'M', 6)
    if inBounds(r+1, c-1):
        count += traverse(r+1, c-1, 'M', 1)
    if inBounds(r+1, c):
        count += traverse(r+1, c, 'M', 2)
    if inBounds(r+1, c+1):
        count += traverse(r+1, c+1, 'M', 3)
    
    return count

###

total = 0
for i in range(num_r):
    for j in range(num_c):
        if grid[i][j] == 'X':
            total += start(i, j)

print(total)
```

```python
ilename = 'test2.txt'
filename = 'input.txt'

oppMap = {'M': 'S',
          'S': 'M'}

grid = []
with open(filename, 'r') as f:
    for line in f:
        grid.append(list(line.strip()))

num_r = len(grid)
num_c = len(grid[0])

###

def checkCorners(r:int, c:int) -> bool:
    return ((0 <= r-1 < num_r) and (0 <= r+1 < num_r)) and \
            ((0 <= c-1 < num_c) and (0 <= c+1 < num_c))

def check(r:int, c:int):
    if not checkCorners(r, c):
        return 0
    
    tl = oppMap.get(grid[r-1][c-1], 0)
    br = oppMap.get(grid[r+1][c+1], 0)
    tr = oppMap.get(grid[r-1][c+1], 0)
    bl = oppMap.get(grid[r+1][c-1], 0)

    if (tl == 0) or (br == 0) or (tr == 0) or (bl == 0):
        return 0
    elif (tl != oppMap.get(br)) or (tr != oppMap.get(bl)):
        return 0
    else:
        return 1
    
###

total = 0
for i in range(num_r):
    for j in range(num_c):
        if grid[i][j] == 'A':
            total += check(i, j)

print(total)
```
