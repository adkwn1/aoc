```python
import sys

filename = './input.txt'
sys.setrecursionlimit(int(10e7))

with open(filename, 'r') as f:
    garden = [line.strip() for line in f]
    
num_r, num_c = len(garden), len(garden[0])
seen = set()

def inBounds(r, c, s):
    return (0 <= r < num_r) and (0 <= c < num_c) and garden[r][c] == s

def canMoveUp(r, c, s):
    return inBounds(r-1, c, s) and (r-1, c) not in seen

def canMoveRight(r, c, s):
    return inBounds(r, c+1, s) and (r, c+1) not in seen

def canMoveDown(r, c, s):
    return inBounds(r+1, c, s) and (r+1, c) not in seen

def canMoveLeft(r, c, s):
    return inBounds(r, c-1, s) and (r, c-1) not in seen

def canMove(r, c, s):
    return canMoveUp(r, c, s) or \
            canMoveRight(r, c, s) or \
            canMoveDown(r, c, s) or \
            canMoveLeft(r, c, s)
           
def calcPerimeter(r, c, s):
    perimeter = 0
    if not inBounds(r-1, c, s):
        perimeter += 1
    if not inBounds(r, c+1, s):
        perimeter += 1
    if not inBounds(r+1, c, s):
        perimeter += 1
    if not inBounds(r, c-1, s):
        perimeter += 1
    return perimeter

def dfs(row, col, species):
    seen.add((row, col))
       
    if not canMove(row, col, species):
        perimeter = calcPerimeter(row, col, species)
        return (1, perimeter)

    else:
        area = 1
        perimeter = calcPerimeter(row, col, species)
        if canMoveUp(row, col, species):
            a, p = dfs(row-1, col, species)
            area += a
            perimeter += p
        if canMoveRight(row, col, species):
            a, p = dfs(row, col+1, species)
            area += a
            perimeter += p
        if canMoveDown(row, col, species):
            a, p = dfs(row+1, col, species)
            area += a
            perimeter += p
        if canMoveLeft(row, col, species):
            a, p = dfs(row, col-1, species)
            area += a
            perimeter += p
        return area, perimeter

cost = 0
for i in range(num_r):
    for j in range(num_c):
        if (i, j) not in seen:
            species = garden[i][j]
            a, p = dfs(i, j, species)
            cost += a*p

print(cost)
```

```python
import sys

filename = './input.txt'
sys.setrecursionlimit(int(10e7))

with open(filename, 'r') as f:
    garden = [line.strip() for line in f]
    
num_r, num_c = len(garden), len(garden[0])
seen = set()
r_sides = {}
c_sides = {}
def inBounds(r, c, s):
    return (0 <= r < num_r) and (0 <= c < num_c) and garden[r][c] == s

def canMoveUp(r, c, s):
    return inBounds(r-1, c, s) and (r-1, c) not in seen

def canMoveRight(r, c, s):
    return inBounds(r, c+1, s) and (r, c+1) not in seen

def canMoveDown(r, c, s):
    return inBounds(r+1, c, s) and (r+1, c) not in seen

def canMoveLeft(r, c, s):
    return inBounds(r, c-1, s) and (r, c-1) not in seen

def canMove(r, c, s):
    return canMoveUp(r, c, s) or \
            canMoveRight(r, c, s) or \
            canMoveDown(r, c, s) or \
            canMoveLeft(r, c, s)
           
def calcSides(s):
    count = 0
    for r, col in r_sides.items():
        top_on = bot_on = False
        prev = None
        for c in sorted(col):
            if not inBounds(r-1, c, s) and (not top_on or c-1 != prev):
                count += 1
                top_on = True
            elif inBounds(r-1, c, s):
                top_on = False
            
            if not inBounds(r+1, c, s) and (not bot_on or c-1 != prev):
                count += 1
                bot_on = True
            elif inBounds(r+1, c, s):
                bot_on = False
            prev = c
    
    for c, row in c_sides.items():
        lt_on = rt_on = False
        prev = None
        for r in sorted(row):
            if not inBounds(r, c-1, s) and (not lt_on or r-1 != prev):
                count += 1
                lt_on = True
            elif inBounds(r, c-1, s):
                lt_on = False
            
            if not inBounds(r, c+1, s) and (not rt_on or r-1 != prev):
                count += 1
                rt_on = True
            elif inBounds(r, c+1, s):
                rt_on = False
            prev = r
    return count

def dfs(row, col, species):
    seen.add((row, col))
    r_sides[row] = r_sides.get(row, []) + [col]
    c_sides[col] = c_sides.get(col, []) + [row]
    if not canMove(row, col, species):
        return 1

    else:
        area = 1
        if canMoveUp(row, col, species):
            area += dfs(row-1, col, species)
        if canMoveRight(row, col, species):
            area += dfs(row, col+1, species)
        if canMoveDown(row, col, species):
            area += dfs(row+1, col, species)
        if canMoveLeft(row, col, species):
            area += dfs(row, col-1, species)

        return area

cost = 0
for i in range(num_r):
    for j in range(num_c):
        if (i, j) not in seen:
            r_sides = {}
            c_sides = {}
            species = garden[i][j]
            a = dfs(i, j, species)
            p = calcSides(species)
            cost += a*p

print(cost)
```
