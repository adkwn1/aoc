```python
filename = './input.txt'
#filename = './test.txt'

with open(filename, 'r') as f:
    nums = f.read().split('\n')

def morph(n):
    n = (n*64)^n
    n = n%16777216
    n = (n//32)^n
    n = n%16777216
    n = (n*2048)^n
    n = n%16777216

    return n

monkeys = {}
for buyer, n in enumerate(nums):
    monkeys[buyer] = [int(n[-1])]
    for i in range(2000):
        n = morph(int(n))
        monkeys[buyer].append(int(str(n)[-1]))
```

```python
size = 3
memo = {}
for k, v in monkeys.items():
    i = 1
    seen = set()
    while i + size < len(v):
        a = v[i] - v[i-1]
        b = v[i+1] - v[i]
        c = v[i+2] - v[i+1]
        d = v[i+3] - v[i+2]

        if (a, b, c, d) not in memo:
            memo[(a, b, c, d)] = v[i+3]
        elif (a, b, c, d) in memo and (a, b, c, d) not in seen:
            memo[(a, b, c, d)] += v[i+3]
        
        seen.add((a, b, c, d))
        i += 1
```
