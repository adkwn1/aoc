```python
import sys
filename = './input.txt'

with open(filename, 'r') as f:
    arr = [int(x) for x in f.readline().split()]

sys.setrecursionlimit(int(10e7))

seen = {}
def blink(s, count):
    if (s, count) in seen:
        return seen.get((s, count))
    if count == 0:
        newstones = 1
    elif s == 0:
        newstones = blink(1, count-1)
    elif not len(str(s)) % 2:
        s = str(s)
        left = int(s[:len(s)//2])
        right = int(s[len(s)//2:])
        newstones = blink(left, count-1) + blink(right, count-1)
    else:
        newstones = blink(s*2024, count-1)
    
    seen[(s, count)] = newstones
    return newstones

total = 0
for s in arr:
    total += blink(s, 75)

print(total)
```
