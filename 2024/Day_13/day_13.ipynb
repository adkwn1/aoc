```python
import re

filename = './input.txt'
#filename = './test.txt'

def solve(ax, ay, bx, by, px, py):   
    A = (by*px - bx*py) / (ax*by - bx*ay)
    B = (py*ax - px*ay) / (ax*by - bx*ay)

    if abs(A - round(A)) < 0.0001 and abs(B - round(B)) < 0.0001:
        cost = 3*A + B
    else:
        cost = 0

    return cost

total = 0
with open(filename, 'r') as f:
    d = f.read().strip()
    for idx, game in enumerate(d.split('\n\n')):
        ax, ay, bx, by, px, py = [int(x) for x in re.findall(r'\d+', game)]
        px += 10**13
        py += 10**13
        total += solve(ax, ay, bx, by, px, py)

print(int(total))
```
