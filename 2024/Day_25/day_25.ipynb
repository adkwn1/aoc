```python
import sys

sys.setrecursionlimit(int(10e7))

filename = './input.txt'
#filename = './test.txt'

keys = {}
locks = {}
num_c = 5
num_r = 7
with open(filename, 'r') as f:
    k = 0
    l = 0
    for chunk in f.read().split('\n\n'):
        if chunk[0] == '#':
            tumbler = [0]*num_c
            for line in chunk.split('\n'):
                for c in range(num_c):
                    if line[c] == '#':
                        tumbler[c] += 1
            locks[l] = tumbler
            l += 1

        if chunk[0] == '.':
            pins = [0]*num_c
            for line in chunk.split('\n'):
                for c in range(num_c):
                    if line[c] == '#':
                        pins[c] += 1
            keys[k] = pins
            k += 1

count = 0
for tumblers in locks.values():
    for pins in keys.values():
        fits = True
        for i in range(num_c):
            if (tumblers[i] + pins[i]) > 7:
                fits = False
                break
        if fits:
            count += 1

print(count)
```
