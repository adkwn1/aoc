```python
#filename = './test.txt'
filename = './input.txt'

left = []
right = []

with open(filename, 'r') as f:
    for line in f:
        left.append(int(line.split()[0]))
        right.append(int(line.split()[1]))

left.sort()
right.sort()

total = 0
for i in range(len(left)):
    total += abs(left[i] - right[i])

print(total)
```

```python
from collections import Counter

#filename = './test.txt'
filename = './input.txt'

left = Counter()
right = Counter()

with open(filename, 'r') as f:
    for line in f:
        l = int(line.split()[0])
        r = int(line.split()[1])
        left.update([l])
        right.update([r])

total = 0

for k, v in left.items():
    total += k * v * right.get(k, 0)

print(total)
```
