```python
filename = './test.txt'
#filename = './input.txt'


with open(filename, 'r') as f:
    disk = [int(x) for x in f.readline()]


fileMap = []
fileId = 0
for i in range(len(disk)):
    if i%2:
        fileMap.append(0)
    else:
        fileMap.append(fileId)
        fileId += 1

def moveFiles(disk:list, left:int, right:int):
    rem = disk[left] - disk[right]
    file = disk[right]
    disk[left] = 0
    disk.insert(left+1, rem)
    disk.insert(left+1, file)
    disk[right+1] += disk.pop(right+2)
    if right+2 < len(disk):
        disk[right+1] += disk.pop(right+2)
    
    fid = fileMap[right]
    fileMap[left] = 0
    fileMap.insert(left+1, 0)
    fileMap.insert(left+1, fid)
    fileMap.pop(right+2)
    if right+2 < len(fileMap):
        fileMap.pop(right+2)

left = 1
right = len(disk) - 1
while left < right:
    i = left
    j = right
    
    while i < j:
        if disk[i] >= disk[j]:
            moveFiles(disk, i, j)
            if i == left:
                while disk[left] != 0:
                    left += 2
            break
        else:
            i += 2
    right -= 2

checksum = 0
block = 0
for i in range(len(disk)):
    if not i%2:
        for j in range(disk[i]):
            checksum += block*fileMap[i]
            block += 1
    elif disk[i] != 0:
        for j in range(disk[i]):
            block += 1

print(checksum)
# 6285437315929 too low
# 9706905316018 too high
```
