```python
#filename = './test.txt'
filename = './input.txt'

with open(filename, 'r') as f:
    disk = [int(num) for num in f.readline()]

i  = 0
j = len(disk)-1
block = 0
left_id = 0
right_id = len(disk)//2
checksum = 0
space = 0

while i < j+1:
    if space > 0:
        for k in range(min(space, disk[j])):
            checksum += block*right_id 
            block += 1
        if space >= disk[j]:
            space -= disk[j]
            j -= 2
            right_id -= 1
        else:
            disk[j] -= space
            space = 0
    elif i%2:
        space = disk[i]
        i += 1
    else:
        size = disk[i]
        for k in range(size):
            checksum += block*left_id
            block += 1
        left_id += 1
        i += 1

print(checksum)
```

```python
#filename = './test.txt'
filename = './input.txt'

disk = {}
file_id = 0
idx = 0
with open(filename, 'r') as f:
    for i, size in enumerate(f.readline()):
        if i % 2:
            for rem in range(int(size), 0, -1):
                disk.update({idx: ('.', rem)})
                idx += 1
        else:
            for k in range(int(size)):
                disk.update({idx: (file_id, int(size))})
                idx += 1
            file_id += 1

left = 0 + disk.get(0)[1]
right = len(disk) - 1

while left < right:
    i = left
    j = right
    while i < j:
        space = disk.get(i)[1]
        fid, size = disk.get(j)
        if space >= size:
            if j == (len(disk)-1) or disk.get(j+1)[0] != '.':
                newspace = 1
            else:
                newspace = disk.get(j+1)[1] + 1
            # move and vacate
            for k in range(size):
                disk.update({i: (fid, size)})
                disk.update({j: ('.', newspace)})
                j -= 1
                i += 1
                newspace += 1
            # fix vacated space sizes
            while disk.get(j)[0] == '.':
                disk.update({j: ('.', newspace)})
                j -= 1
                newspace += 1
            break            
        else:
            i += space
            while disk.get(i)[0] != '.':
                i += 1
    # either moved or exhausted all possibilities
    # update left and or right
    while disk.get(right)[0] != fid-1:
        right -= 1
    while disk.get(left)[0] != '.':
        left += 1

checksum = 0
for block, file in disk.items():
    if file[0] != '.':
        checksum += block*file[0]

print(checksum)
```
