```python
#filename = './test.txt'
filename = './input.txt'

grid = []
chars = '<^>v'
found = False
i = 0
with open(filename, 'r') as f:
    for line in f:
        row = line.strip()
        grid.append(row)

        if not found:
            for j in range(len(row)):
                if (row[j] != '#') and (row[j] != '.'):
                    found = True
                    start = (i, j)

        i += 1

num_r = len(grid)
num_c = len(grid[0])
visited = set()
count = 0

def moveNorth(r, c):
    global count
    while True:
        if ((r, c) not in visited):
            visited.add((r, c))
            count += 1
        if r-1 < 0:
            return 'E', r, c
        else:
            if grid[r-1][c] == '#':
                return '>', r, c
            else:
                r -= 1
        
def moveEast(r, c):
    global count
    while True:
        if ((r, c) not in visited):
            visited.add((r, c))
            count += 1
        if c+1 >= num_c:
            return 'E', r, c
        else:
            if grid[r][c+1] == '#':
                return 'v', r, c
            else:
                c += 1

def moveSouth(r, c):
    global count
    while True:
        if ((r, c) not in visited):
            visited.add((r, c))
            count += 1
        if r+1 >= num_r:
            return 'E', r, c
        else:
            if grid[r+1][c] == '#':
                return '<', r, c
            else:
                r += 1
    
def moveWest(r, c):
    global count
    while True:
        if ((r, c) not in visited):
            visited.add((r, c))
            count += 1
        if c-1 < 0:
            return 'E', r, c
        else:
            if grid[r][c-1] == '#':
                return '^', r, c
            else:
                c -= 1

moves = {'^': moveNorth,
         '>': moveEast,
         'v': moveSouth,
         '<': moveWest}

r, c = start
guard = grid[r][c]

while guard != 'E':
    guard, r, c = moves[guard](r, c)

print(count)
```
