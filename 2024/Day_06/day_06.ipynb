```python
#filename = './test.txt'
filename = './input.txt'

grid = []
found = False
i = 0
with open(filename, 'r') as f:
    for line in f:
        row = line.strip()
        grid.append(row)

        if not found:
            for j in range(len(row)):
                if (row[j] != '#') and (row[j] != '.'):
                    found = True
                    start = (i, j)

        i += 1

num_r = len(grid)
num_c = len(grid[0])

dirCoord = {'^': (-1, 0),
            '>': (0, 1), 
            'v': (1, 0), 
            '<': (0, -1)}

dirNext = {'^': '>',
           '>': 'v',
           'v': '<',
           '<': '^'}

new_blocks = set()
attempted = {start}

def inBounds(r, c):
    return (0<=r<num_r) and (0<=c<num_c)

def move(dir, r, c, visited=set(), block=None, detect=False):
    while True:
        if (dir, r, c) in visited:
            new_blocks.add(block)
            return
        else:
            visited.add((dir, r, c))
            
        a, b = dirCoord.get(dir)
        if not inBounds(r+a, c+b):
            return
        else:
            if grid[r+a][c+b] == '#' or (r+a, c+b) == block:
                dir = dirNext.get(dir)
            else:
                if not detect:
                    if (r+a, c+b) not in attempted:
                        attempted.add((r+a, c+b))
                        local = visited.copy()
                        move(dirNext.get(dir), r, c, local, (r+a, c+b), True)
                r += a
                c += b

r, c = start
guard = grid[r][c]
move(guard, r, c)
print(len(new_blocks))
```
