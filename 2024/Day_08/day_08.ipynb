```python
from itertools import combinations

#filename = './test.txt'
filename = './input.txt'

grid = []
ant_coords = {}
with open(filename, 'r') as f:
    x = 0
    for line in f:
        row = line.strip()
        grid.append(row)
        for y in range(len(row)):
            if row[y] != '.':
                ant_coords[row[y]] = ant_coords.get(row[y], []) + [(x,y)]
        x += 1

num_r = len(grid)
num_c = len(grid[0])
seen = set()

def inBounds(r, c):
    return (0 <= r < num_r) and (0 <= c < num_c)

def findAntinodes(coords:list):
    for comb in combinations(coords, 2):
        x, y = comb[0]
        a, b = comb[1]
        
        #same column
        if x == a:
            r = x
            c = max(y, b)
            while inBounds(r, c):
                if (r, c) not in seen:
                    seen.add((r, c))
                c += 1

            c = min(y, b)    
            while inBounds(r, c):
                if (r, c) not in seen:
                    seen.add((r, c))
                c -= 1

        #same row
        elif y == b:
            r = max(x, a)
            c = y
            while inBounds(r, c):
                if (r, c) not in seen:
                    seen.add((r, c))
                r += 1

            r = min(x, a)    
            while inBounds(r, c):
                if (r, c) not in seen:
                    seen.add((r, c))
                r -= 1

        else:
            r = x
            c = y
            while inBounds(r, c):
                if (r, c) not in seen:
                    seen.add((r, c))
                r += x-a
                c += y-b
            
            r = a
            c = b
            while inBounds(r, c):
                if (r, c) not in seen:
                    seen.add((r, c))
                r += a-x
                c += b-y
    
for k, v, in ant_coords.items():
    findAntinodes(v)

print(len(seen))
```
