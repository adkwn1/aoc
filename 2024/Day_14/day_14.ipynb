```python
filename = './input.txt'
#filename = './test.txt'

num_r = 103
num_c = 101 
pos = {}
vels = {}
with open(filename, 'r') as f:
    for idx, line in enumerate(f.readlines()):
        p, v = line.split()
        pc, pr = p.split('p=')[1].split(',')
        pc, pr = int(pc), int(pr)
        vc, vr = v.split('v=')[1].split(',')
        vc, vr = int(vc), int(vr)
        pos.update({idx: (pr, pc)})
        vels.update({idx: (vr, vc)})


def updateR(r):
    if 0 <= r < num_r:
        return r
    else:
        return r % num_r

def updateC(c):
    if 0 <= c < num_c:
        return c
    else:
        return c % num_c

def printTree():
    coords = sorted(set(pos.values()))
    tree = []
    for i in range(num_r):
        row = []
        for j in range(num_c):
            if (i, j) in coords:
                row.append('X')
            else:
                row.append('.')
        tree.append(row)
        
    for r in tree:
        print(r)

def checkTree():
    start = False
    max_length = 0
    count = 0
    for i in range(num_r):
        for j in range(num_c):
            if start and (i, j) in pos.values():
                count += 1
            elif start and (i, j) not in pos.values():
                start = False
                if count > max_length:
                    max_length = count
                count = 0
            elif (i, j) in pos.values():
                start = True
    return max_length > 10

def simulate(ff, time):
    for idx, coords in pos.items():
        r = updateR(coords[0] + vels[idx][0]*ff)
        c = updateC(coords[1] + vels[idx][1]*ff)
        pos.update({idx:(r, c)})            

    count = ff + 1
    while count <= time:
        for idx, coords in pos.items():
            r = updateR(coords[0] + vels[idx][0])
            c = updateC(coords[1] + vels[idx][1])
            
            pos.update({idx:(r, c)})
            
        if checkTree():
            printTree()
            return count

        count += 1

time = simulate(6869,6871)
print(time)
```
