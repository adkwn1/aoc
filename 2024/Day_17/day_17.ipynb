```python
filename = './input.txt'
#filename = './test.txt'

register = {}
program = []
combo = {0: 0,
         1: 1,
         2: 2,
         3: 3}

with open(filename, 'r') as f:
    reg, prog = f.read().split('\n\n')
    for key, r in enumerate(reg.split('\n')):
        left, right = r.split(':')
        val = int(right.strip())
        register.update({key: val})
        
    for p in prog.split(':')[1].strip().split(','):
        program.append(int(p))
```

```python
def run(program, a, b, c):
    combo.update({4: a})
    combo.update({5: b})
    combo.update({6: c})

    i = 0
    while i < len(program)-1:
        opcode = program[i]
        operand = program[i+1]
               
        if opcode == 0:
            combo[4] = int(combo[4] / 2**combo[operand])
        elif opcode == 1:
            combo[5] = combo[5] ^ operand
        elif opcode == 2:
            combo[5] = combo[operand] % 8
        elif opcode == 3:
            if combo[4] != 0:
                i = operand - 2
        elif opcode == 4:
            combo[5] = combo[5]^combo[6]
        elif opcode == 5:
            yield combo[operand] % 8
        elif opcode == 6:
            combo[5] = int(combo[4] / 2**combo[operand])
        elif opcode == 7:
            combo[6] = int(combo[4] / 2**combo[operand])
        
        i += 2

print(', '.join(map(str, run(program, register[0], register[1], register[2]))))
```

```python
def expect(prog_in, exp_out, prev_a=0):
    if not exp_out:
        return prev_a
    for a in range(1 << 10):
        if a >> 3 == prev_a & 127 and next(run(prog_in, a, 0, 0)) == exp_out[-1]:
            ret = expect(prog_in, exp_out[:-1], (prev_a << 3) | (a % 8))
            if ret is not None:
                return ret
            
print(expect(program, program))
```
