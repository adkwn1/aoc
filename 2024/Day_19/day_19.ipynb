```python
import re
import sys

filename = './input.txt'
sys.setrecursionlimit(int(10e7))

with open(filename, 'r') as f:
    a, b = f.read().split('\n\n')
    pats = a.split(', ')
    desg = b.split('\n')

seen = {}
def dfs(rem:str):
    if rem == '':
        return 1
    elif rem in seen:
        return seen[rem]
    else:
        count = 0
        for i in range(1, len(rem)+1):
            if rem[:i] in pats:
                count += dfs(rem[i:])
        
        seen[rem] = count
        return count

count = 0
for d in desg:
    count += dfs(d)

print(count)
```
