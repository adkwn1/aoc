```python
#filename = './test.txt'
filename = './input.txt'

equations = {}
with open(filename, 'r') as f:
    for line in f:
        line = line.split(':')
        equations.update({line[0]: line[1].split()})

def solver(value:int, idx:int, rem:list):
    if idx == 0:
        return value == int(rem[idx])
    else:
        num = int(rem[idx])
        idx -= 1
        if value % num == 0:            
            return solver(value//num, idx, rem) or solver(value - num, idx, rem)
        else:
            return solver(value - num, idx, rem)

count = 0
for k, v in equations.items():
    if solver(int(k), len(v)-1, v):
        count += int(k)

print(count)
```

```python
#filename = './test.txt'
filename = './input.txt'

equations = {}
with open(filename, 'r') as f:
    for line in f:
        line = line.split(':')
        equations.update({line[0]: line[1].split()})

def solver(value:int, total:int, rem:list):
    if (total == value) and (not rem):
        return True
    elif (total > value) or (not rem):
        return False
    else:
        plus = total + int(rem[0])
        star = total * int(rem[0])
        concat = int(str(total) + rem[0])
        return solver(value, plus, rem[1:]) or solver(value, star, rem[1:]) or solver(value, concat, rem[1:])

count = 0
for k, v in equations.items():
    if solver(int(k), int(v[0]), v[1:]):
        count += int(k)

print(count)
```
