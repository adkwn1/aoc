```python
import heapq
filename = './input.txt'
#filename = './test.txt'

num_r = num_c = 71
#num_r = num_c = 7
grid = []
for i in range(num_r):
    row = []
    for j in range(num_c):
        row.append('.')
    grid.append(row)

with open(filename, 'r') as f:
    count = 0
    for line in f.readlines():
        c, r = line.strip().split(',')
        grid[int(r)][int(c)] = '#'
        count += 1
        if count == 2914:
            break

steps = {1: (-1, 0),
         2: (0, 1),
         3: (1, 0),
         4: (0, -1)}

def inBounds(r, c):
    return 0 <= r < num_r and 0 <= c < num_c

seen = set()
dist = {}
min_dist = None
queue = []
r, c = (0, 0)
end = (70, 70)
heapq.heappush(queue, (0, r, c))
while queue:
    cost, r, c = heapq.heappop(queue)
    if (r, c) not in dist:
        dist[(r, c)] = cost
    if (r, c) == end and not min_dist:
        min_dist = cost
    if (r, c) in seen:
        continue
    
    seen.add((r, c))
    # straight
    for d, (nr, nc) in steps.items():
        rr = r + nr
        cc = c + nc

        if inBounds(rr, cc) and grid[rr][cc] != '#':
            heapq.heappush(queue, (cost + 1, rr, cc))
        
print(min_dist)
```
