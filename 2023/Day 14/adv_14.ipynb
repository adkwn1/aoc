In short: if you move the rocks, you can focus the dish. The platform even has a control panel on the side that lets you tilt it in one of four directions! The rounded rocks (O) will roll when the platform is tilted, while the cube-shaped rocks (#) will stay in place. You note the positions of all of the empty spaces (.) and rocks (your puzzle input). The amount of load caused by a single rounded rock (O) is equal to the number of rows from the rock to the south edge of the platform, including the row the rock is on. (Cube-shaped rocks (#) don't contribute to load.) 

```python
data = []
with open('input.txt') as file:
    for line in file:
        data.append(list(line.strip()))

num_r = len(data)
num_c = len(data[0])

weight = 0
for c in range(num_c):
    h = 0
    p = 0
    while p < num_r:
        if data[p][c] == '.':
            p += 1
        elif data[p][c] == 'O':
            data[p][c] = '.'
            data[h][c] = 'O'
            weight += num_r - h
            h += 1
            p += 1
        elif data[p][c] == '#':
            p += 1
            h = p

print(weight)
```

Each cycle tilts the platform four times so that the rounded rocks roll north, then west, then south, then east. After each tilt, the rounded rocks roll as far as they can before the platform tilts in the next direction. After one cycle, the platform will have finished rolling the rounded rocks in those four directions in that order. This process should work if you leave it running long enough, but you're still worried about the north support beams. To make sure they'll survive for a while, you need to calculate the total load on the north support beams after 1000000000 cycles.

```python
from copy import deepcopy

#cycle = n, w, s, e
def tilt_n(d, row, col):
    for c in range(col):
        h = 0
        p = 0
        while p < row:
            if d[p][c] == '.':
                p += 1
            elif d[p][c] == 'O':
                d[p][c] = '.'
                d[h][c] = 'O'
                h += 1
                p += 1
            elif d[p][c] == '#':
                p += 1
                h = p

def tilt_w(d, row, col):
    for r in range(row):
        h = 0
        p = 0
        while p < col:
            if d[r][p] == '.':
                p += 1
            elif d[r][p] == 'O':
                d[r][p] = '.'
                d[r][h] = 'O'
                h += 1
                p += 1
            elif d[r][p] == '#':
                p += 1
                h = p

def tilt_s(d, row, col):
    for c in range(col):
        h = row - 1 
        p = row - 1
        while p >= 0:
            if d[p][c] == '.':
                p -= 1
            elif d[p][c] == 'O':
                d[p][c] = '.'
                d[h][c] = 'O'
                h -= 1
                p -= 1
            elif d[p][c] == '#':
                p -= 1
                h = p

def tilt_e(d, row, col):
    for r in range(row):
        h = col - 1
        p = col - 1
        while p >= 0:
            if d[r][p] == '.':
                p -= 1
            elif d[r][p] == 'O':
                d[r][p] = '.'
                d[r][h] = 'O'
                h -= 1
                p -= 1
            elif d[r][p] == '#':
                p -= 1
                h = p

def calc_load(d, row, col):
    weight = 0
    for r in range(row):
        for c in range(col):
            if d[r][c] == 'O':
                weight += row - r
    return weight

def hash_image(d):
    s = ''
    for row in d:
        temp = ''.join(row)
        s += temp
    return s

data = []
with open('input.txt') as file:
    for line in file:
        data.append(list(line.strip()))

num_r = len(data)
num_c = len(data[0])

i = 1
num_cycles = 1e9
memo = {}
weights = {}
while i <= num_cycles:
    tilt_n(data, num_r, num_c)
    tilt_w(data, num_r, num_c)
    tilt_s(data, num_r, num_c)
    tilt_e(data, num_r, num_c)
    key = hash_image(data)
    w = calc_load(data, num_r, num_c)
    if key in memo:
        start = memo[key]
        end = i - 1
        break
    else:
        memo[key] = i
        weights[i] = w
    i += 1

cyc_len = end - start + 1
idx = (num_cycles - start)%cyc_len + start

print(weights[idx])
```
