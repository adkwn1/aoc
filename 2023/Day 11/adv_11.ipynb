The researcher has collected a bunch of data and compiled the data into a single giant image (your puzzle input). The image includes empty space (.) and galaxies (#). The researcher is trying to figure out the sum of the lengths of the shortest path between every pair of galaxies. However, there's a catch: the universe expanded in the time it took the light from those galaxies to reach the observatory. Due to something involving gravitational effects, only some space expands. In fact, the result is that any rows or columns that contain no galaxies should all actually be twice as big. In this example, after expanding the universe, the sum of the shortest path between all 36 pairs of galaxies is 374.

Expand the universe, then find the length of the shortest path between every pair of galaxies. What is the sum of these lengths?

```python
cosmos = []
with open('input.txt') as file:
    for line in file:
        s = line.strip()
        cosmos.append(s)
        if '#' not in s:
            cosmos.append(s)

num_rows = len(cosmos)
num_cols = len(cosmos[0])
transpose = []
coordinates = {}
t_cols = 0
galaxy_count = 0
for c in range(num_cols):
    s = []
    count = 0
    for r in range(num_rows):
        if cosmos[r][c] == '#':
            galaxy_count += 1
            s.append(str(galaxy_count))
            count += 1
            coordinates[galaxy_count] = (t_cols, r)
        else:
            s.append(cosmos[r][c])
    transpose.append(s)
    t_cols += 1
    if count == 0:
        transpose.append(s)
        t_cols += 1

dist = 0
for i in range(1, galaxy_count):
    for j in range(i+1, galaxy_count+1):
        a = coordinates[i]
        b = coordinates[j]
        dist += abs(b[0] - a[0]) + abs(b[1] - a[1])

print(dist)
```

The galaxies are much older (and thus much farther apart) than the researcher initially estimated.

Now, instead of the expansion you did before, make each empty row or column one million times larger. That is, each empty row should be replaced with 1000000 empty rows, and each empty column should be replaced with 1000000 empty columns.

```python
cosmos = []
empty_rows = []
empty_cols = []
with open('input.txt') as file:
    row_num = 0
    for line in file:
        s = line.strip()
        cosmos.append(s)
        if '#' not in s:
            empty_cols.append(row_num)
        row_num += 1

num_rows = len(cosmos)
num_cols = len(cosmos[0])
transpose = []
coordinates = {}
galaxy_count = 0
for c in range(num_cols):
    s = []
    count = 0
    for r in range(num_rows):
        if cosmos[r][c] == '#':
            galaxy_count += 1
            s.append(str(galaxy_count))
            count += 1
            coordinates[galaxy_count] = (c, r)
        else:
            s.append(cosmos[r][c])
    transpose.append(s)
    if count == 0:
        empty_rows.append(c)

dist = 0
for i in range(1, galaxy_count):
    for j in range(i+1, galaxy_count+1):
        a = coordinates[i]
        b = coordinates[j]

        min_r = min(b[0], a[0])
        max_r = max(b[0], a[0])
        min_c = min(b[1], a[1])
        max_c = max(b[1], a[1])

        for r in empty_rows:
            if min_r < r and r < max_r:
                dist += 999999
        for c in empty_cols:
            if min_c < c and c < max_c:
                dist += 999999
        dist += abs(b[0] - a[0]) + abs(b[1] - a[1])
        
print(dist)
```
