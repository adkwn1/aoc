You have nowhere near that many stars - you need to find a way to disconnect at least half of the equipment here, but it's already Christmas! You only have time to disconnect three wires.
Fortunately, someone left a wiring diagram (your puzzle input) that shows how the components are connected. Find the three wires you need to disconnect in order to divide the components into two separate groups. What do you get if you multiply the sizes of these two groups together?

```python
import networkx as nx

g = nx.Graph()
with open('input.txt') as file:
    for line in file:
        u, nodes = line.split(':')
        for v in nodes.split():
            g.add_edge(u, v)
            g.add_edge(v, u)

g.remove_edges_from(nx.minimum_edge_cut(g))
a, b = nx.connected_components(g)

print(len(a)*len(b))
```
