After examining the maps a bit longer, your attention is drawn to a curious fact: the number of nodes with names ending in A is equal to the number ending in Z! If you were a ghost, you'd probably just start at every node that ends with A and follow all of the paths at the same time until they all simultaneously end up at nodes that end with Z. Simultaneously start on every node that ends with A. How many steps does it take before you're only on nodes that end with Z?

```python
import re
from functools import reduce
import math

node_map = {}
seq_map = {'L': 0, 'R': 1}

with open('input.txt') as file:
    seq = file.readline().strip()
    for line in file:
        s = line.split('=')       
        key = s[0].strip()
        path = re.sub(r'[\(,\)]', '', s[1].strip()).split()        
        node_map[key] = path

node_list = []
for k in node_map.keys():
    if k[-1] == 'A':
        node_list.append(k)

def get_steps(node):
    i = 0
    steps = 0
    length = len(seq)
    while True:
        if node[-1] == 'Z':
            break
        if i == length:
            i = 0
        if seq[i] == 'L':
            node = node_map[node][0]
        else:
            node = node_map[node][1]
        steps += 1
        i += 1
    
    return steps

results = []
for n in node_list:
    results.append(get_steps(n))
    
def lcm(numbers):
    return reduce(lambda x, y: x * y // math.gcd(x, y), numbers, 1)
 
print(lcm(results))
```
